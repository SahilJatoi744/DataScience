# Assignment Solution

```python
# Q1
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("netflix_data.csv")
data.head()
```

```python
def cast_len(val):
    if val=="None":
        return 0
    else:
        return len(str(val).split(", "))
```

```python
data["cast"].isnull().sum()
```

```python
data["cast"].replace(np.nan,"None",inplace=True)
```

```python
data.tail()
```

```python
data["Cast_length"] = data["cast"].apply(cast_len)

data.head()
```

```python
data.tail()
```
