# Scatter Plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x,y,'o',color='black')
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y,'-ok',color='black')
plt.show()
```

```python
plt.plot(x, y, '-p', color='black')
plt.show()
```

```python
x = np.random.randn(100) # randn does normal distribution by gaussion distribution
y = np.random.randn(100)
colors = np.random.randn(100)

size = 1000*np.random.rand(100)

plt.scatter(x,y,c=colors,s=size,alpha=0.3) # alpha make the colors transparent
plt.colorbar()
plt.show()
```

# Histogram

```python
data = np.random.randn(1000)
plt.hist(data)
```

```python
#bins zoom in(-) and out(+)
plt.hist(data,bins=30,alpha=0.5,histtype='stepfilled',color='steelblue')
plt.show()
```

```python
# USed in ratio of covid etc
x1 = np.random.normal(0,0.8,1000)
x2 = np.random.normal(-2,1,1000)
x3 = np.random.normal(3,2,1000)

dict_extra = dict(histtype='stepfilled',alpha=0.3,bins=30)
# ** --> to pass dictionary and * --> to pass as a list 
plt.hist(x1, **dict_extra)
plt.hist(x2, **dict_extra)
plt.hist(x3, **dict_extra)
plt.show()
```

# Styling

```python
x = np.linspace(0,10,100)

plt.plot(x,np.sin(x),'-b',label='Sine Graph')
plt.plot(x,np.cos(x),'--r',label='Cosine Graph')

plt.show()
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), '-b', label='Sine Graph')
plt.plot(x, np.cos(x), '--r', label='Cosine Graph')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), '-b', label='Sine Graph')
plt.plot(x, np.cos(x), '--r', label='Cosine Graph')
plt.legend(loc='upper left')
plt.show()
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), '-b', label='Sine Graph')
plt.plot(x, np.cos(x), '--r', label='Cosine Graph')
plt.legend(loc='lower left')
plt.show()
```

```python
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), '-b', label='Sine Graph')
plt.plot(x, np.cos(x), '--r', label='Cosine Graph')
plt.legend(loc='best')
plt.show()
```

```python
# Hear Beats
x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), '-b', label='Sine Graph')
plt.plot(x, np.cos(x), '--r', label='Cosine Graph')
plt.legend(loc='best',ncol=2)
plt.show()
```

## 2D PLOTS

```python
# X-rays
x,y = np.random.multivariate_normal([0,0],[[1,1],[1,2]],1000).T

plt.hist2d(x,y,bins=20,cmap='Blues')
plt.show()
```
