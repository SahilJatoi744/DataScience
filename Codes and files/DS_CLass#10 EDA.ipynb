# Scatter Plot


```python
# Importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black')
```

```python
plt.plot(x, y, '-ok', color='black')
```

```python
plt.scatter(x, y)
```

# Histogram

```python
data = np.random.randn(1000)
print(len(data))
print(data)
plt.hist(data, bins=40, histtype="stepfilled")
# bins --> + increse so it will zoom out and - decrease so it will zoom in
plt.show()
```

```python
x = np.linspace(0, 10, 1000)
inp = input("Enter function Sin or Cos")
if inp == "Sin":
    plt.plot(x, np.sin(x), label="Sine")
else:
    plt.plot(x, np.cos(x), label="Cosine")

plt.legend()

```

```python
plt.plot(x, np.sin(x), label="Sine")
plt.plot(x, np.cos(x), label="Cosine")
# ncol number of columns
# fancybox is for legend box shape true means round edges
plt.legend(loc="lower left", ncol=2, fancybox=True, borderpad=1, shadow=True)
```
